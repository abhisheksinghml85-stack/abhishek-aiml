# Student Performance Analysis

This project analyzes student exam performance data using Python and EDA techniques.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

Load Dataset

```python
df = pd.read_csv('/content/StudentsPerformance (1).csv')
df.sample(8)
```

Understand Dataset

```python
df.shape
df.describe()
df.info()
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

Visualization

```python
sns.histplot(df['math score'],kde= 'True')
plt.xlabel('Math Score')
plt.ylabel('Frequency')
plt.title('Distribution of Math Scores')
plt.show()
```

Test Preparation vs Scores

```python
plt.bar(df['test preparation course'], df['math score'])
plt.xlabel('Test Preparation Course')
plt.ylabel('Math Score')
plt.title('Effect of Test Preparation on Math Score')
plt.show()
```

```python
sns.histplot(df['math score'], bins=10,color='blue', edgecolor='black',alpha=0.4,kde= 'True')
sns.histplot(df['reading score'], bins=10,color='red', edgecolor='black',alpha=0.7,kde= 'True')
plt.xlabel('Math Score')
plt.ylabel('Frequency')
plt.title('Distribution of Math Scores and Reading Scores side by side')
plt.legend(['Math Scores', 'Reading Scores'])
plt.show()
```

Feature Engineering

```python
df['average_score'] = (
    df['math score']
    + df['reading score']
    + df['writing score']
)/3
```
